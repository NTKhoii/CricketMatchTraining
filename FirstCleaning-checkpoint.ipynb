```python
#split up Span into Start and End Date
df['Span'].str.split(pat='-')
```

```python
df['Rookie_Year']=df['Span'].str.split(pat='-').str[0]
```

```python
df['Final_Year']=df['Span'].str.split(pat='-').str[1]
```

```python
df
```

```python
df=df.drop(columns='Span')
```

```python
df
```

```python
#split the country out of the name
df['Player'].str.split(pat=['()'])
df['Name']=df['Player'].str.split(pat='(').str[0]
df['Country']=df['Player'].str.split(pat='(').str[1]
```

```python
df
```

```python
df['Country'].str.split(')')
df['Country']=df['Country'].str.split(')').str[0]
```

```python
df
```

```python
df.dtypes
```

```python
df['highest_inns_score']=df['highest_inns_score'].str.split(pat='*').str[0]
```

```python
df['highest_inns_score']=df['highest_inns_score'].astype('int')
```

```python
#change the type of Rookie_Year and Final_Year to int
df[['Rookie_Year','Final_Year']]=df[['Rookie_Year','Final_Year']].astype('int')
```

```python
df['Mat']=df['Mat'].str.split('*').str[0]
df['Mat']=df['Mat'].astype('int')
```

```python
df.dtypes
```

```python
df['balls_faced']=df['balls_faced'].str.split('+').str[0]
```

```python
df['balls_faced']=df['balls_faced'].fillna(0)
```

```python
df['balls_faced']=df['balls_faced'].astype('int')
```

```python
df.dtypes
```

```python
## Build career_length columns
df['Career_length']=df['Final_Year']-df['Rookie_Year']
```

```python
## Calculation with the dataframe
```

```python
#Ques1: Cricketers in this DF what is the Average career length
df['Career_length'].mean()
```

```python
#Ques2: AVG Batting_Strike_Rate for cricketers who played over 10 years
df[df['Career_length']>10]['batting_strike_rate'].mean()
```

```python
#Ques3: find the number of cricketers who played before 1960
df[df['Rookie_Year']<1960]['Player'].count()
```

```python
#Ques4: Max height Inns Score by country
df.groupby('Country')['highest_inns_score'].max().to_frame('HighInCountry').reset_index().sort_values('HighInCountry')
```
